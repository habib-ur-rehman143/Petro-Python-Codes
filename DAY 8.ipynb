# visualization
- matplotlib

```python
import matplotlib.pyplot as plt
```

```python
sw = np.linspace(0,1,101)
```

```python
sw
```

```python
Krw = sw**3
```

```python
Krw
```

```python
Kro = (1-sw)**2
```

```python
Kro
```

- step 1 Krw vs sw plot

```python
plt.plot(sw,Krw)
```

- step 2 Kro vs sw plot

```python
plt.plot(sw, Kro)
```

- step 3  Krw vs Kro on same plot

```python
plt.plot(sw,Krw)
plt.plot(sw, Kro)
```

```python
# step 4 give x axis label
plt.plot(sw,Krw)
plt.plot(sw, Kro)

plt.xlabel('water saturation fraction')
plt.ylabel('relative permeability (fraction)')
```

```python
# step 6 grid
plt.plot(sw,Krw)
plt.plot(sw, Kro)

plt.xlabel('water saturation fraction')
plt.ylabel('relative permeability (fraction)')
plt.grid(True)
```

```python
# step 7 : adding lagends
plt.plot(sw,Krw, label='Krw(water rel pem)')
plt.plot(sw, Kro, label='Kro(oil rel pem)')

plt.xlabel('water saturation fraction')
plt.ylabel('relative permeability (fraction)')
plt.grid(True)
plt.legend()
```

```python
# step 8 changing the size of plot
plt.figure(figsize=(12,8))
plt.plot(sw,Krw, label='Krw(water rel pem)')
plt.plot(sw, Kro, label='Kro(oil rel pem)')

plt.xlabel('water saturation fraction')
plt.ylabel('relative permeability (fraction)')
plt.grid(True)
plt.legend()
```

```python
# step 9: changing the line colour
plt.figure(figsize=(12,8))
plt.plot(sw,Krw, label='Krw(water rel pem)', color="blue")
plt.plot(sw, Kro, label='Kro(oil rel pem)', color="brown")

plt.xlabel('water saturation fraction')
plt.ylabel('relative permeability (fraction)')
plt.grid(True)
plt.legend()
```

```python
#step 10 : line thickness
plt.figure(figsize=(12,8))
plt.plot(sw,Krw, label='Krw(water rel pem)', color="blue", lw=0.6)
plt.plot(sw, Kro, label='Kro(oil rel pem)', color="brown", lw=4)

plt.xlabel('water saturation fraction')
plt.ylabel('relative permeability (fraction)')
plt.grid(True)
plt.legend()
```

```python
# step 11 scatter plot
plt.figure(figsize=(12,8))
plt.plot(sw,Krw, label='Krw(water rel pem)', color="blue", lw=0.6)
plt.scatter(sw,Krw)
plt.plot(sw, Kro, label='Kro(oil rel pem)', color="brown", lw=4)
plt.scatter(sw,Kro)
plt.xlabel('water saturation fraction')
plt.ylabel('relative permeability (fraction)')
plt.grid(True)
plt.legend()
```

```python
#step 12 title of plot
plt.figure(figsize=(12,8))
plt.plot(sw,Krw, label='Krw(water rel pem)', color="blue", lw=0.6)
plt.scatter(sw,Krw)
plt.plot(sw, Kro, label='Kro(oil rel pem)', color="brown", lw=4)
plt.scatter(sw,Kro)
plt.xlabel('water saturation fraction')
plt.ylabel('relative permeability (fraction)')
plt.grid(True)
plt.legend()
plt.title('Relative Permeability Curve')
```

```python
plt.style.available
```

```python
# step 13: making my plot stylish
plt.style.use('classic')
plt.figure(figsize=(12,8))
plt.plot(sw,Krw, label='Krw(water rel pem)', color="blue", lw=0.6)
plt.scatter(sw,Krw)
plt.plot(sw, Kro, label='Kro(oil rel pem)', color="brown", lw=4)
plt.scatter(sw,Kro)
plt.xlabel('water saturation fraction')
plt.ylabel('relative permeability (fraction)')
plt.grid(True)
plt.legend()
plt.title('Relative Permeability Curve')
```

```python
# step 14: limit x and y axis
plt.style.use('classic')
plt.figure(figsize=(12,8))
plt.plot(sw,Krw, label='Krw(water rel pem)', color="blue", lw=0.6)
plt.scatter(sw,Krw)
plt.plot(sw, Kro, label='Kro(oil rel pem)', color="brown", lw=4)
plt.scatter(sw,Kro)

plt.xlabel('water saturation fraction')
plt.xlim(0,1)
plt.ylabel('relative permeability (fraction)')
plt.ylim(0,1)

plt.grid(True)
plt.legend()
plt.title('Relative Permeability Curve')
```

```python
# final step for saving the plot
plt.style.use('classic')
plt.figure(figsize=(12,8))
plt.plot(sw,Krw, label='Krw(water rel pem)', color="blue", lw=0.6)
plt.scatter(sw,Krw)
plt.plot(sw, Kro, label='Kro(oil rel pem)', color="brown", lw=4)
plt.scatter(sw,Kro)

plt.xlabel('water saturation fraction')
plt.xlim(0,1)
plt.ylabel('relative permeability (fraction)')
plt.ylim(0,1)

plt.grid(True)
plt.legend()
plt.title('Relative Permeability Curve')

plt.savefig('relative_perm_curve')
```
